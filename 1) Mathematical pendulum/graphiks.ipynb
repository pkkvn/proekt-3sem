```python
import matplotlib.pyplot as plt;
import numpy as np;

#траектория в фазовом пространстве для численного решения

time, phi, xi = np.loadtxt('res1.txt', unpack=True)

fig, (g1, g2, g3) = plt.subplots(1, 3)

             
g1.scatter(phi, xi, 1)
g2.scatter(time, phi, 1)
g3.scatter(time, xi, 1)
      
g1.set_xlabel('v')
g1.set_ylabel('x')    
g2.set_xlabel('v')
g3.set_xlabel('v')

plt.show()
print (g1)
```

```python
#траектория в фазовом пространстве для точного решения

time, phi, xi = np.loadtxt('res2.txt', unpack=True)

fig, (g1, g2, g3) = plt.subplots(1, 3)

             
g1.scatter(phi, xi, 1)
g2.scatter(time, phi, 1)
g3.scatter(time, xi, 1)

g1.set_xlabel('v')
g1.set_ylabel('x')    
g2.set_xlabel('v')
g3.set_xlabel('v')
             
plt.show()
```

```python
#Проверка закона сохранения энергии для численного решения
time, phi, xi = np.loadtxt('res1.txt', unpack=True)

fig, (g1) = plt.subplots(1, 1)

             
g1.scatter(time, xi*xi/2 + phi*phi/2, 1)
g1.set_xlabel('time')
g1.set_ylabel('energy')
plt.grid()
plt.show()
```
